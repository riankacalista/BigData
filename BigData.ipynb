```python
from minio import Minio
from minio.error import S3Error
import pandas as pd
import sqlalchemy
import io

client = Minio(
    "localhost:9000",
    access_key="minioadmin", 
    secret_key="minioadmin123", 
    secure=False
)

# Buat bucket kalau belum ada
bucket = "minilab-datalake"
if not client.bucket_exists(bucket):
    client.make_bucket(bucket)
    print(f"✅ Bucket '{bucket}' berhasil dibuat!")
else:
    print(f"✅ Bucket '{bucket}' sudah ada!")

# Fungsi upload DataFrame ke MinIO
def upload_to_minio(df, path):
    csv_bytes = df.to_csv(index=False).encode('utf-8')
    client.put_object(
        bucket,
        path,
        data=io.BytesIO(csv_bytes),
        length=len(csv_bytes),
        content_type="text/csv"
    )
    print(f"🚀 Berhasil upload ke: {path} ({len(df)} baris)")

# 2. Koneksi PostgreSQL
engine = sqlalchemy.create_engine('postgresql://admin:password123@localhost:5432/minilab_db')

# --- BAGIAN A: Ingestion dari RDBMS ---
print("\n--- Processing RDBMS Data ---")
# Menarik data dari tabel E-Commerce Anda
df_customers = pd.read_sql("SELECT * FROM customers", engine)
df_products  = pd.read_sql("SELECT * FROM products", engine)
df_orders_db = pd.read_sql("SELECT * FROM orders", engine)

upload_to_minio(df_customers, "raw/rdbms/customersrdbms.csv")
upload_to_minio(df_products,  "raw/rdbms/productsrdbms.csv")
upload_to_minio(df_orders_db, "raw/rdbms/ordersrdbms.csv")

# --- BAGIAN B: Ingestion dari File Lokal (Dataset Berbeda) ---
print("\n--- Processing File Data ---")

# 1. Ingestion dari CSV (Menggunakan file orders_source.csv yang baru Anda buat)
try:
    df_csv = pd.read_csv("products.csv")
    upload_to_minio(df_csv, "raw/csv/products.csv")
except FileNotFoundError:
    print("❌ Error: File 'products.csv' tidak ditemukan!")

# 2. Ingestion dari XLSX (Opsional: Jika Anda punya file excel lain)
# Jika belum ada file .xlsx, Anda bisa skip bagian ini atau buat dummy singkat
try:
    df_xlsx = pd.read_excel("orders_source.xlsx") # Ganti nama file jika ada
    upload_to_minio(df_xlsx, "raw/xlsx/orders.csv")
except Exception as e:
    print(f"⚠️ Ingestion XLSX dilewati karena: {e}")

print("\n Semua data berhasil di-ingestion ke MinIO!")
```

```python
from minio import Minio
import pandas as pd
import sqlalchemy
import io

# ========================
# 1. Koneksi ke MinIO
# ========================
client = Minio(
    "localhost:9000",
    access_key="minioadmin", 
    secret_key="minioadmin123", 
    secure=False
)

bucket = "minilab-datalake"

if not client.bucket_exists(bucket):
    client.make_bucket(bucket)
    print(f"✅ Bucket '{bucket}' berhasil dibuat!")
else:
    print(f"✅ Bucket '{bucket}' sudah ada!")

# ========================
# 2. Fungsi Upload
# ========================
def upload_to_minio(df, path):
    csv_bytes = df.to_csv(index=False).encode('utf-8')
    client.put_object(
        bucket,
        path,
        data=io.BytesIO(csv_bytes),
        length=len(csv_bytes),
        content_type="text/csv"
    )
    print(f"🚀 Upload: {path} ({len(df)} baris)")

# ========================
# 3. Koneksi PostgreSQL (FIXED)
# ========================
engine = sqlalchemy.create_engine(
    'postgresql://admin:admin123@localhost:5432/minilab_db'
)

# ========================
# 4. Ingestion dari Database
# ========================
print("\n--- Processing RDBMS Data ---")

df_customers = pd.read_sql("SELECT * FROM customers", engine)
df_products  = pd.read_sql("SELECT * FROM products", engine)
df_orders_db = pd.read_sql("SELECT * FROM orders", engine)

upload_to_minio(df_customers, "raw/rdbms/customers.csv")
upload_to_minio(df_products,  "raw/rdbms/products.csv")
upload_to_minio(df_orders_db, "raw/rdbms/orders.csv")

# ========================
# 5. Ingestion dari CSV
# ========================
print("\n--- Processing CSV Data ---")

try:
    df_csv = pd.read_csv("products.csv")
    upload_to_minio(df_csv, "raw/csv/products.csv")
except FileNotFoundError:
    print("❌ File products.csv tidak ditemukan")

# ========================
# 6. Ingestion dari Excel
# ========================
print("\n--- Processing Excel Data ---")

try:
    df_xlsx = pd.read_excel("orders_source.xlsx")
    upload_to_minio(df_xlsx, "raw/xlsx/orders.xlsx")
except Exception as e:
    print(f"⚠️ Skip XLSX: {e}")

# ========================
# DONE
# ========================
print("\n Semua data berhasil masuk ke Data Lake (MinIO)")
```
